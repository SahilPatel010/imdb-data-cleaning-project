```python
string_df['country'].value_counts()
```

```python
# Fill missing values with 'Other'
string_df['country'] = string_df['country'].fillna('Other')

# Replace rare countries (less than 50 occurrences) with 'Other'
rare_countries = string_df['country'].value_counts()[string_df['country'].value_counts() < 50].index
string_df['country'] = string_df['country'].replace(rare_countries, 'Other')

# Check unique values after preprocessing
print(string_df['country'].unique())
```

```python
string_df["country"].isna().sum()
```

```python
countries_dummies = pd.get_dummies(string_df['country'], prefix='country',drop_first=True).astype(int)
string_df = pd.concat([string_df, countries_dummies], axis=1)
string_df= string_df.drop('country', axis=1)
```

```python

string_df["content_rating"].isna().sum()
```

```python
# Fill missing values with 'Not Rated'
string_df['content_rating'] = string_df['content_rating'].fillna('Not Rated')

# Combine rare or uncommon ratings into 'Other'
rare_ratings = string_df['content_rating'].value_counts()[string_df['content_rating'].value_counts() < 50].index
string_df['content_rating'] = string_df['content_rating'].replace(rare_ratings, 'Other')

# Check unique values after preprocessing
print(string_df['content_rating'].unique())
```

```python
content_rating_dummies= pd.get_dummies(string_df['content_rating'], prefix='content_rating',drop_first=True).astype(int)
string_df = pd.concat([string_df, content_rating_dummies], axis=1)
string_df= string_df.drop('content_rating', axis=1)
```
